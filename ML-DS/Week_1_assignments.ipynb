```python
#homework 3 (using an algorithm on the dataset)

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from pandas.plotting import parallel_coordinates
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df.groupby('variety').size()
```

```python
train, test = train_test_split(df,
                               test_size = 0.4,
                               stratify=df['variety'],
                               random_state=42)
```

```python
n_bins = 10
fig, axs = plt.subplots(2, 2)
axs[0,0].hist(train['sepal.length'], bins = n_bins);
axs[0,0].set_title('Sepal Length');
axs[0,1].hist(train['sepal.width'], bins = n_bins);
axs[0,1].set_title('Sepal Width');
axs[1,0].hist(train['petal.length'], bins = n_bins);
axs[1,0].set_title('Petal Length');
axs[1,1].hist(train['petal.width'], bins = n_bins);
axs[1,1].set_title('Petal Width');

fig.tight_layout(pad=1.0);
```

```python
#plotting a scatterplot of all-paired attributes

sns.pairplot(train, hue="variety", height = 2, palette = 'colorblind');
```

```python
#making a correlation matrix to examine relationship between variables

corrmat = train.corr()
sns.heatmap(corrmat, annot = True, square = True);
```

```python
#separating class label and features

X_train = train[['sepal.length','sepal.width','petal.length','petal.width']]
y_train = train.variety
X_test = test[['sepal.length','sepal.width','petal.length','petal.width']]
y_test = test.variety
```

```python
#building a classification tree

mod_dt = DecisionTreeClassifier(max_depth = 3, random_state = 1)
mod_dt.fit(X_train,y_train)
prediction=mod_dt.predict(X_test)
```

```python
print("the accuracy is ""{:.3}".format(metrics.accuracy_score(prediction,y_test)))
```
